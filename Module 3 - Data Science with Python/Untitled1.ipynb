```python
from matminer.datasets import load_dataset

df = load_dataset("jarvis_dft_3d")
df
```

```python
df.describe()
```

```python
# target output column
K = df['elasticity.K_VRH'].values
G = df['elasticity.G_VRH'].values
E = 9*K*G/(3*K + G)

# possible descriptor columns
excluded = ["elasticity.K_VRH","elasticity.G_VRH", "H", "pretty_formula",
            "volume", "nsites", "spacegroup.symbol", "e_above_hull", "composition"]
X = df.drop(excluded, axis=1)

X["elastic_modulus"] = E
X.describe()
X.to_csv("elastic_modulus.csv", index=False)
```
