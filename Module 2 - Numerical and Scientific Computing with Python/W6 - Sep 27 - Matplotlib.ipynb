## Creating and customizing simple plots

`pyplot` provides a procedural interface to the matplotlib object-oriented plotting library.

```python
x = range(10)
y = [2*item for item in x]
plt.plot(x, y)
```

```python
x = np.arange(10)
y = 2*x
plt.plot(x, y);
```

No need for a `hold on` to plot multiple lines in the same figure.

```python
x = np.linspace(0, 2*np.pi, 30)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1)
plt.plot(x, y2);
```

**Tick labels**

**Labels and titles: `xlabel`, `ylabel`, and `title`**

## Figures, Subplots and Axes objects

A **Figure** in matplotlib represents the whole window, which can include multiple plots (subplots).

For simple plots, we can quickly create the figure and axes *implicity* (as we did above). However, the figure, subplots, and axes can also be controlled *explicitly*, giving you more control of the plots.

- **subplots** position the plots in a rectangular grid, by specifying the rows and columns.
- **axes** allows free placement of plots at any location in the figure.

Of course, you can have multiple subplots and axes in a single figure.

Weve already worked with figures and subplots (without explicitly calling them) by calling `plot`. In the background, matplotlib calls `gca()` to get the current axes and `gca` in turn calls `gcf()` to get the current figure. If there is `none` it calls `subplots()` to make one.

![figure anatomy](https://matplotlib.org/stable/_images/anatomy.png)

Empty figure with no Axes:

```python
fig = plt.figure()
```

A figure with a single axis:

```python
fig, ax = plt.subplots()  # a figure with a single Axes
```

Plot on the axis:

```python
ax.plot(x, y1)
```

Do it together to plot inline:

```python
fig, ax = plt.subplots()  # a figure with a single Axes
ax.plot(x, y1);
```

### Customizing plots with these objects

Create a figure with a single subplot, and define a custom Figure size:

```python
fig, ax = plt.subplots(figsize=(8, 5))
```

Plot with an OOP paradigm:

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y1, "bs--", label="Sine")
ax.plot(x, y2, "r-.", label = "Cosine")

ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
              [r"$0$", r"$\pi/2$", r"$\pi$", r"$3\pi/2$", r"$2\pi$"])
ax.set_yticks([-1, 0, 1])
ax.set_xlim(0, 2*np.pi)

ax.set_xlabel(r"Angle, $\theta$ (radians)")
ax.set_ylabel("Function")
ax.set_title("Sine and Cosine")

ax.grid(True)

ax.legend();
```

**Moving spines**

```python
# Previous plot
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y1, "bs--", label="Sine")
ax.plot(x, y2, "r-.", label = "Cosine")
ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
              [r"$0$", r"$\pi/2$", r"$\pi$", r"$3\pi/2$", r"$2\pi$"])
ax.set_xlim(0, 2*np.pi)
ax.set_yticks([-1, 0, 1])
ax.grid(True)
ax.legend()


# Remove top and right spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Move bottom spine to y=0
ax.spines['bottom'].set_position(('data',0))

# Left spine is left in the default position
```

**Additional aesthetics**

```python
# Previous plot
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y1, "bs--", label="Sine")
ax.plot(x, y2, "r-.", label = "Cosine")
ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
              [r"$0$", r"$\pi/2$", r"$\pi$", r"$3\pi/2$", r"$2\pi$"])
ax.set_yticks([-1, 0, 1])
ax.set_xlim(0, 2*np.pi)
ax.grid(True)
ax.legend()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))


for xlabel, ylabel in zip(ax.get_xticklabels(), ax.get_yticklabels()):
    ylabel.set_fontsize(16)
    xlabel.set_fontsize(16)
    xlabel.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.8))
```

### Subplots

```python
fig, axs = plt.subplots(1, 2, figsize=(16, 5), layout="constrained")
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16, 5), layout="constrained")

# First subplot
axs[0].plot(x,y1, "bs--")
axs[0].set_ylabel(r"sin($\theta$)")

# Second subplot
axs[1].plot(x,y2, "r-.")
axs[1].set_ylabel(r"cos($\theta$)")

# Common to both subplots
for ax in axs:
    ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
                  [r"$0$", r"$\pi/2$", r"$\pi$", r"$3\pi/2$", r"$2\pi$"])
    ax.set_yticks([-1, 0, 1])
    ax.set_xlim(0, 2*np.pi)
    ax.set_xlabel(r"$\theta$ (radians)")
    ax.grid(True)
```

You can make improve this by using a function

```python
def ax_plotter(ax, x, y, color):
    """Helper function for subplots."""
    ax.plot(x,y, color=color)
    
    ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
                  [r"$0$", r"$\pi/2$", r"$\pi$", r"$3\pi/2$", r"$2\pi$"])
    ax.set_yticks([-1, 0, 1])
    ax.set_xlim(0, 2*np.pi)
    ax.set_xlabel(r"$\theta$ (radians)")
    ax.grid(True)
    
    return ax

fig, axs = plt.subplots(1, 2, figsize=(16, 5))

y_list = [y1, y2]
colors = ["b", "r"]

for ax, y, color in zip(axs, y_list, colors):
    ax = ax_plotter(ax, x, y, color)
```

**Indexing subplots**

```python
fig, axs = plt.subplots(ncols=3, nrows=2, figsize=(16, 5), layout="constrained")

for i_row, row in enumerate(axs):
    for i_col, col in enumerate(row):
        col.annotate(f"{i_row}, {i_col}",
                     (0.5, 0.5), ha="center", va="center",
                     fontsize=36)
```

Another way to plot on specific subplots is to acccess the specific item in the numpy array:

```python
fig, axs = plt.subplots(ncols=3, nrows=2, figsize=(16, 5), layout="constrained")

axs[1, 0].annotate(f"Second row,\nFirst column",
                   (0.5, 0.5), ha="center", va="center",
                   fontsize=24);
```
